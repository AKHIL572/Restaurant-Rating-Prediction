```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# ------------------------------
# Paths
# ------------------------------
PROJECT_ROOT = Path.cwd().parent
DATA_PATH = PROJECT_ROOT / "Dataset" / "Res_dataset.csv"
OUTPUT_PATH = PROJECT_ROOT / "Dataset" / "cleaned_dataset.csv"
```

```python
# ------------------------------
# Load Dataset
# ------------------------------
df = pd.read_csv(DATA_PATH)
print("Dataset Loaded Successfully.\n")
```

```python
df.head()
```

```python
print("Shape of the dataset :", df.shape)
```

```python
df.columns.tolist()
```

```python
print("\nDuplicate Rows:", df.duplicated().sum())
print("Duplicate Restaurant IDs:", df.duplicated(subset="Restaurant ID").sum())
```

```python
# Remove duplicate Restaurant IDs if any
df.drop_duplicates(subset="Restaurant ID", inplace=True)
```

```python
df.info()
```

```python
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
# Fill missing Cuisines
df["Cuisines"] = df["Cuisines"].fillna("Unknown")
```

```python
# ------------------------------
# Convert Yes/No Columns
# ------------------------------
yes_no_cols = [
    "Has Table booking",
    "Has Online delivery",
    "Is delivering now",
    "Switch to order menu"
]

df[yes_no_cols] = df[yes_no_cols].replace({"Yes": 1, "No": 0}).astype(int)
```

```python
# ------------------------------
# Remove Constant Columns
# ------------------------------
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
print("\nConstant Columns:", constant_cols)

df.drop(columns=constant_cols, inplace=True)
```

```python
# ------------------------------
# Convert Categorical Numerical Columns
# ------------------------------
df["Country Code"] = df["Country Code"].astype("category")
df["Price range"] = df["Price range"].astype("category")
```

```python
# ------------------------------
# Unique Value Analysis
# ------------------------------
print("\nUnique Value Counts:\n")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

```python
# ------------------------------
# Target Variable Understanding
# ------------------------------
print("\nTarget Variable: Aggregate rating")
print("Min:", df["Aggregate rating"].min())
print("Max:", df["Aggregate rating"].max())
print("Mean:", df["Aggregate rating"].mean())
print("Median:", df["Aggregate rating"].median())

print("\nNumber of Restaurants with 0 Rating:",
      (df["Aggregate rating"] == 0).sum())
```

```python
# Statistical Summary
df.describe()
```

```python
df.to_csv(OUTPUT_PATH, index = False)
```
